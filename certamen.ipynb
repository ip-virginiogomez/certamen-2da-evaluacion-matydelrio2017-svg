### Ejercicio 3.1 – Calculadora de propina (12 pts)

Escribe un programa que:
1. Pida el **total de la cuenta** (puede ser decimal).

3. Calcule la **propina** y el **total a pagar**.

**Salida esperada:**
```
Ingresa el total de la cuenta: 8500
Ingresa el porcentaje de propina (%): 10
----------------------------
Propina: $850.0
Total a pagar: $9350.0
```


````python
# Ejercicio 3.1 – Calculadora de propina
# Escribe tu código aquí:

cuenta=int(input("Ingrese el total de cuenta: "))
propina=int(input("Ingrese un porcentaje de propina: "))
propinapago=cuenta/propina
totalpago=cuenta+propinapago
print(f"Propina: {propinapago}")
print(f"Total a pagar: {totalpago}")
````

### Ejercicio 3.2 – Clasificador de notas (16 pts)

Escribe un programa que:

3. Muestre el **concepto** y si está **aprobado o reprobado**.

| Rango        | Concepto      |
|--------------|---------------|
| 6.0 – 7.0    | Excelente     |
| 5.0 – 5.9    | Bueno         |
| 4.0 – 4.9    | Suficiente    |
| 1.0 – 3.9    | Insuficiente  |

**Salida esperada:**
```

Concepto: Bueno
Estado: Aprobado
```


````python
# Ejercicio 3.2 – Clasificador de notas
# Escribe tu código aquí:
nota=int(input("Ingrese nota entre 1.0 y 7.0: "))

if nota<=39: 
    print("Concepto: Insuficiente")
    print("Estado: Reprobado")
if nota==40 & nota<=49:
    print("Concepto: Suficiente")
    print("Estado: Aprobado")
if nota==50 & nota<=59:
    print("Concepto: Bueno")
    print("Estado: Aprobado")
if nota>=60:
    print("Concepto: ¡EXCELENTE!")
    print("Estado: Aprobado")

````
